```python
import pandas as pd
import numpy as np
from PyPDF2 import PdfFileReader
import PyPDF2
import requests
import io
from bs4 import BeautifulSoup
import tabula
import os
import pdfminer
import pdfminer.high_level
import html5lib
```

```python
fp = open('abanca1.pdf', 'rb')
pdfFile = PdfFileReader(fp)
if pdfFile.isEncrypted:
    try:
        pdfFile = pdfFile.decrypt('')
        print('File Decrypted (PyPDF2)')
    except:
        command = ("cp "+ 'abanca1.pdf' +
            " temp.pdf; qpdf --password='' --decrypt temp.pdf " + 'abanca1.pdf'
            + "; rm temp.pdf")
        os.system(command)
        print('File Decrypted (qpdf)')
        fp = open('abanca1.pdf', 'rb')
        pdfFile = PdfFileReader(fp)
else:
    print('File Not Encrypted')
```

```python
from io import StringIO
from pdfminer.high_level import extract_text_to_fp
from pdfminer.layout import LAParams
output = StringIO()
with open('ctt.pdf', 'rb') as pdf_file:
    extract_text_to_fp(pdf_file, output, laparams=LAParams(), output_type='html', codec=None)
with open('ctt.html', 'a') as html_file:
    html_file.write(output.getvalue())
```
